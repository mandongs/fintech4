# if else 조건문

## 1. 조건문? 특정 조건을 판별해서 작업을 분기시키는 문법
```python
if 조건식:
    조건이 참일 때 실행할 코드
else:
    조건이 거짓일 때 실행할 코드
```

````python
 x = 12
````

````python
# x가 10인지 아닌지 판별하는 조건문
if x == 10 and x % 2 == 0:
    print(f"{x}는, 10입니다.")
else:
    print(f"{x}는, 10이 아닙니다.")
````

* 숫자를 입력받아 홀수인지 짝수인지 판단하는 프로그램

````python
x = input("숫자를 입력하세요: ")
print(x, type(x))
if int(x) % 2 == 0:
    print(f"입력한 숫자{x}는 짝수입니다.")
else:
    print(f"입력한 숫자{x}는 홀수입니다.")
````

* if문의 조건을 여러개 주기

````python
# 숫자를 입력받아서 숫자가 1-10 사이의 숫자인지 판별하고 
# 홀수인지 짝수인지 판별해서
# 입력받은 숫자 {x}는 1-10 사이의 숫자이고/숫자가 아니고 홀수/짝수입니다.

x = input("숫자를 입력해주세요: ")
x = int(x)
# x가 1-10 사이의 숫자이고 짝수인 경우
if 1 <= x <= 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자이고 짝수입니다.")
else: 
    print(f"{x}는 1-10 사이의 숫자가 아니고 홀수입니다.")
````

## 2. 다중 조건문 사용하기 elif
``` python
if 조건:
    조건이 참일 때 실행할 코드
elif 조건2:
    조건 2가 참일 때 실행할 코드
elif 조건3:
    조건 3가 참일 때 실행할 코드
elif:
    위의 조건 1,2,3이 모두 거짓일 때 실행할 코드
```

````python
y = int(input("숫자를 입력하세요: "))
print(y, type(y))
if 1 <= y <= 50:
#    print(f'{y}는 1-50사이의 숫자입니다."
    if y % 3 == 0:
        print("3의 배수입니다.")
    else:
        print("3의 배수가 아닙니다.")
else:
    print("1-50 사이의 정수로 입력해 주세요.")
````

# 3. 조건 표현식
```
참일 때 실행할 코드 if 조건 else 거짓일 때 실행할 코드
```

````python
# 일반적인 if else 문으로 홀수, 짝수 구분하는 코드
num = int(input("숫자를 입력하세요"))
print(num, type(num))

if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
````

````python
# 조건표현식 홀수, 짝수 구분하는 코드
num = int(input("숫자를 입력하세요"))
print(num, type(num))

print("짝수") if num % 2 == 0 else print("홀수")
````
