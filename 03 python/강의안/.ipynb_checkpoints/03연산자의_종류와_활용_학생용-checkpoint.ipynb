* 산술, 할당, 비교, 논리, 식별, 멤버

# 산술연산자 +, -, *, /, %, //, **

```python
a = 10
b = 2
print(a + b)
```

# 할당연산자
* =, +=, -=, *=, /=, %=

# 비교연산자
* 같다 == 
* 다르다 != 
* 초과 >  
* 미만 <  
* 이상 >=
* 이하 <= 

* == 같으면 True, 다르면 False

* != 다르면 True, 같으면 False

# 논리 연산자 
* and, or, not

* and는 두 값이 모두 참이어야 참

or 두 값중 하나라도 참이면 참, 두 값 모두가 거짓이어야 거짓

not 값을 반대로 뒤집음 True=>False, False => True

* and, or, not의 우선순위
* not, and, or순으로 판단

# 식별연산자
* is, is not (object, data type)이 같은지 비교

# 멤버연산자
* in , not in
