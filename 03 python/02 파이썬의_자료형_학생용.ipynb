# 파이썬의 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소숫점이 없는 숫자)
* float: 실수(소숫점이 있는 숫자)
* str: 문자/문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음
* tuple: list와 비슷하나 한 번 만들면 수정이 안됨, 모든 자료형을 넣을 수 있음
* dictionary(dict): key : value 쌍으로 이루어진 자료형, 모든 자료형을 넣을 수 있음

# 리스트  list
* 리스트는 [정수, 실수, 문자, 리스트, 튜플 , 디셔너리 ] 모두 넣을 수있다.
* 리스트는 여러 자료가 같이 나열된 형태이므로 순서가 있다.
* 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 가능하다.

```python
l1 = [1, 2.0, '삼', [1,2,3], (4,5,6), {"key":"value"}]
l1
```

# input() : 키보드 입력을 받는 함수
* input("화면에 표시될 메시지")
* 입력 받는 값은 " "으로 구분해서 여러개를 받을 수 있음
* 입력 받는 값은 모두 str 타입의 문자열

```python
input_name = input("이름을 입력해주세요: ")
print(input_name)
```

input 함수를 이용해서 숫자 2개를 입력받아 더한 후 출력하세요.

```python
num1 = input("숫자 1을 입력해주세요: ")
num2 = input("숫자 2를 입력해주세요: ")
print("num1: ", num1, type(num1))
print("num2: ", num2, type(num2))
print("num1 + num2: ", int(num1) + int(num2))
```

* input().split() 으로 여러 데이터를 list 형태로 받기

```python
nums = input("숫자를 5개 공백으로 구분해서 입력하세요: ").split()
print(nums, type(nums))
```
